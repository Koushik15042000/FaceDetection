```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#cv2 is reverse of RGB i.e BGR
frame = cv2.imread("thor.jpg")
```

```python
plt.imshow(frame)
```

```python
rgbimg = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgbimg)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame)
```

```python
face = faces[0]
```

```python
x, y, w, h = face
```

```python
croppedFace = rgbimg[y:y+h, x:x+w]
```

```python
plt.imshow(croppedFace)
```

```python
out = cv2.rectangle(rgbimg, (x,y), (x+w,y+h), (0,255,0), 4)
```

```python
plt.imshow(out)
```
